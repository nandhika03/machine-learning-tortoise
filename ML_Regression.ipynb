# **Regression**

- Regression is used to understand how a dependent variable changes with respect to one or more independent variables
- There are different types of regression including linear regression, logistic regression, multiple linear regression, random forest regressor, etc..
- Regression comes under the supervised part of machine learning
- In simple linear regression, there are only two variables x and y.
- The equation would be y=mx+c, this would be a straight line when plotted.
- The c value is the value of y, when x = 0
- The m value is the value of the slope.
- It gives the answer to the question: when x increases by 1 unit, how much does y increases?
- When there are two or more independent variables, then all of them are taken into account and the plotted figure would become a plane or hyperplane.

### **Assumptions**

- Linearity: the relationship between x and y is linear
- Homoscedascitiy: the residual variance is constant for all independent variables
   if the errors do not have constant variance, then we can 
    - transformation(log,exponential) of the dependent variable
    - weighted least square: assigning weights to independent variables based on their variance
- Normality: The residuals follow a normal distribution
- Independence: The features and their observations are independent of each other

### **Cost function or Sum of residuals**

- The error between the actual value - the predicted values are squared and summed to get the cost function
- The best value for the slope of the line or the best fit line is found by applying gradient descent or convergence function to the cost function.

### **Gradient Descent**

- This is an optimization algorithm that would be applied on the differentiable function (here the cost function) to find the local minimum
- The learning rate is called as alpha
- Its value should not be too high nor too low

### **Ridge Regression**

- The cost function in the linear regression does not always give the best fit line. 
- It either overfits or underfits the training data
- Hence, in ridge regression we include a **penalizing function** - a function that puts another constraint that in-turns optimizes the existing function in a much better way.
- Here we add another term lambda(slope)2 to the cost function
- The lambda would be a non-negative value starting from 1.
- The lambda is a hyper-parameter and its value can be determined by using GridSearchCV
- This added term further aids in the optimization(here minimize the value of slope) of the cost function.
- Here, the co-efficients of the features would be reduced as close to zero, but it smoothens around it and does not touch zero.

### **Lasso Regression (least absolute shrinkage and selection operator)**

- Lasso regression on the other hand, is very useful when there are a lot of features and we need to eliminate a few.
- The lasso regression is best suited for feature selection as it can converge the co-efficient of features exactly to zero, henceforth eliminating them from the equation, giving a more generalized model.
- Here we add another term lambda|slope| to the cost function.
- As the geometric shape of this constraint is diamond shaped with edge, the contour lines can meet exactly at zero, unlike the circle shape that arises with the ridge constraint.

### **Bias and Variance**


- **Bias** : The model picks up very simple assumptions from the data and leaves out the complexity involved in the data and hence does not predict with good accuracy
- **Variance**: The model is very sensitive to the noise and unnecessary patterns in the data, that it takes up all of that instead of the real patter or trend in the data resulting in a model without good accuracy.

### **R square**

- It is same as the co-efficient of determinatin which measures the proportion of variance of the dependent variable that can be explained by the independent variable present in the regression model
- This value is used to check the performance of the model
- The formula is 1 - Squared sum of residuals(unexplained variance or error in actual - predicted value) / sum of squared totals (total variance)
- The value ranges from 0-1
- The more the value is closer to 1, the better the model is

### **Adjusted R sqaure**

- When there is only one independent variable, the R square works
- But, when the number of independent variables increases, the R sqaure value generally would tend to increase
- This is beacuse another feauture certainly decreases the uncertainity.
- The adjusted r sqaure value makes sure that the r sqaure value decreases, if the added feature does not contribute much to the variance of the model.
- The adjusted r sqaure value makes sure that the r sqaure value increases, if the added feature does contribute much to the variance of the model.
- The formula is 1 - (1-R2)(1-n)/(n-k-1)
- N is the number of observations
- K is the number of features or predictors
- The value ranges from 0-1
- The more the value is closer to 1, the better the model is

### **Multicollinearity in linear regression**

- It is when there is a high correlation between two or more independent variables
- Structural multicollinearity is where we introduce it, like X=X and Z=(X)2
- Data Multicollinearity on the other hand, it something that is already present in the data itself.
- Multicollinearity does not affect the predicted values often
- It does not give way to find the statistically significant independent variable
- It makes the regression co-efficients hard to interpret 
- Now, the Variance Inflation Factor can detect multicollinearity in variables
- It ranges from 1 to any value, where 1 means no multicollinearity and 1-5 values indicate moderate multicollinearity
- When the value is more than 5, then it means that we have look into the issue.
- If it causes big issues, then it is good to linearly combine the correlated variables into one or remove the correlated variables.
- Using Principal Component Analysis, a dimensionality reduction technique can also aid in this problem.
- The Ordinary Least Squares is a method used to find the co-efficients in the linear regression

### **Cross validation**

- Inorder to evaluate the performance of an ML model, we use cross validation
- It can take any evaluation metric (accuracy, F-1 score, etc)
- Here, instead of using 80% of the dataset in training and then testing the model using the 20% test set, it involves the training and testing of the model using different parts of the dataset as training and test sets.

 #### **K-Fold Crossvalidation**

  - In K-Fold CV, the dataset is split into K parts, where K-1 parts are used for training and the other part is used for testing.
  - This process is repeated K-times, by changing the training and testing datasets
  - **'cross_val_score'** in sklearn allows you to perform the K-Fold cross validation
  

### **GridSearch CV**

- There are hyperparameters present in every model like the depth of a decision tree, the lambda value of lasso regression, the K value in K-Nearest neighbors classifier
- In the grid of a lot of hyperparameter values, we would need to determine which works very well for a given model.
- We can give a grid or set of hyperparameters to the GridSearchCV, it would fit the model with all of it and tell us which would be the best to use
- Then it also performs Cross Validation of the model associated with the respective hyperparameter to determine that it would be suitable accross varied sets.








```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([2, 3, 4, 5, 6])
y = np.array([3, 5, 7, 9, 11])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(x1, x2, y, color='blue')

# Create a meshgrid for x1 and x2
x1_surf, x2_surf = np.meshgrid(np.linspace(x1.min(), x1.max(), 100), np.linspace(x2.min(), x2.max(), 100))

# Calculate the corresponding y values for the plane
m1, m2, c = 1, 1, 1  # Example coefficients
y_surf = m1 * x1_surf + m2 * x2_surf + c

# Plot the plane
ax.plot_surface(x1_surf, x2_surf, y_surf, color='red', alpha=0.5)

# Labels
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.show()
```
